```python
import zipfile

with zipfile.ZipFile("apps.csv.zip", 'r') as zip_ref:
    zip_ref.extractall(".")

import os
print(os.listdir("."))
```

```python
import pandas as pd
df = pd.read_csv("apps.csv")
df.info()
```

```python
import matplotlib.pyplot as plt
df['Category'].value_counts().plot(kind='bar', figsize=(12,6))
plt.title("App Distribution by Category")
plt.show()
```

```python
df['Rating'].hist(bins=50)
plt.title("App Ratings Distribution")
plt.show()
```

```python
from textblob import TextBlob
df['Sentiment'] = df['Reviews'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
```

```python
import seaborn as sns
sns.scatterplot(x='Installs', y='Rating', hue='Category', data=df)
plt.title("Installs vs Rating by Category")
plt.show()
```
